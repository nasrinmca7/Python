```python
bmi = float(input("Enter the BMI Index:"))
if(bmi < 18.5):
    print("Under weight")
elif(18.5 <= bmi < 24.9):
    print("Normal weight")
elif(25.0 <= bmi < 29.9):
    print("Over weight")
elif(30.0 <= bmi < 34.9):
    print("Very Overweight")
else:
    print("Obese")
```
