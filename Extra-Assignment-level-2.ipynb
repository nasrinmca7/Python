```python
# print 0 to 20 by using range
for i in range(0, 20):
    print(i)
```

```python
# since range starts at 0 by default, we can use this below code as well
for i in range(20):
    print(i)
```

```python
# print range 10 to 20
for i in range(10, 20):
    print(i, end=" ")
```

```python
# Print number of items in the list by using 'len'
List2 = [10, 20, 14, 55, 43, 87, 76]
length = len(List2)
print(List2)
print("Number of item in the List2:\n",length)
```

```python
str = "Artificial Intelligence"
print(str)
for char in str:
    print(char)
```

```python
# print Odd Numbers in the list
arr = [20,10,16,19,25,1,276,188]
print(tuple(arr))
for i in arr:
    num = i%2
    if(num == 1):
        print(i, "is odd")
```

```python
#print Even Numbers in the list
arr = [20,10,16,19,25,1,276,188]
print(tuple(arr))
for i in arr:
    num = i%2
    if(num == 0):
        print(i, "is Even")
```
