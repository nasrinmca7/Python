```python
# Create a function that checks whether the given number is Odd or Even
```

```python
class OddEven:
    def OddEven():
        num = int(input("Enter a number"))
        if(num % 2 == 0):
           print(num, "is Even number")
        else:
           print(num, "is Odd number")
```

```python
OddEven.OddEven()
```

```python
# Create a function that tells elegibility of marriage for male and female according to thei
# r age limit like 21 for male and 18 for female
```

```python
class ElegiblityForMarriage:
    def Elegible():
        gender = input("Enter Gender")
        age = int(input("Enter Age"))
        
        print("Your Gender:", gender)
        print("Your Age:", age)
       
        if((gender.lower() == "female") and (age>=18)):
            print("ELIGIBLE")
        elif((gender.lower() == "male") and (age>=21)):
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")
```

```python
ElegiblityForMarriage.Elegible()
```

```python
#print area and perimeter of triangle using class and functions
```

```python
class triangle:
    def area():
        height = 32
        breadth = 34
        areaOfTriangle = (height*breadth)/2
        return areaOfTriangle
        
    def perimeter():
        height1 = 2
        height2 = 4
        breadth = 4
        perimeterOfTriangle = height1+height2+breadth
        return perimeterOfTriangle
```

```python
print(f"Area: {triangle.area()}")
print(f"Perimeter: {triangle.perimeter()}")
```
